```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
#merging 12month report
files = [file for file in os.listdir("E:\\Project\\Sales\\Pandas-Data-Science-Tasks-master\\SalesAnalysis\\Sales_Data")]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("E:\\Project\\Sales\\Pandas-Data-Science-Tasks-master\\SalesAnalysis\\Sales_Data\\" + file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("All_Data.csv", index=False)
```

```python
#Cleaning DATA

#removing NaN data

nan_df = all_months_data[all_months_data.isna().any(axis=1)]
nan_df.head()

all_months_data = all_months_data.dropna(how='all')
all_months_data.head()
```

```python
#finding and removing OR data
#-finding
#temp_df = all_months_data[all_months_data['Order Date'].str[0:2] == 'Or']
#temp_df.head

#solution
all_months_data = all_months_data[all_months_data['Order Date'].str[0:2] != 'Or']
```

```python
# convert colums to correct tips

all_months_data['Quantity Ordered'] = pd.to_numeric(all_months_data['Quantity Ordered'])
all_months_data['Price Each'] = pd.to_numeric(all_months_data['Price Each'])
```

```python
#Read Updated DataFrame


all_data = pd.read_csv("E:\\Project\\Sales\\All_Data.csv")
all_data.head()




```

```python
#Add a month row
all_months_data['Month'] = all_months_data['Order Date'].str[0:2]
all_months_data['Month'] = all_months_data['Month'].astype('int')
all_months_data.head()
```

```python
#Add Sales Col

all_months_data['Sales'] = all_months_data['Quantity Ordered'] * all_months_data['Price Each']
all_months_data.head()
```

```python
#insert city col
#using .apply()
def get_city(address):
    return address.split(',')[1]

def get_State(address):
    return address.split(',')[2].split(' ')[1]

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x) + ',' + get_State(x))  #fsting can also be used
all_months_data.head()
```

```python
#- Q1. What was the best month for sales? How much was earned that month?

maxSales = all_months_data.groupby('Month').sum()

```

```python

month = range(1,13)

plt.bar(month, maxSales['Sales'])
plt.xticks(month)
plt.xlabel('Month in Number')
plt.ylabel('Sales in $')
plt.show()
```

```python
##What is the city with highiest number of sales

CitySales = all_months_data.groupby('City').sum()
CitySales
```

```python

City = [City for City, df in all_months_data.groupby('City')]

plt.bar(City, CitySales['Sales'])
plt.xticks(City, rotation='vertical', size=5 )
plt.xlabel('City Name')
plt.ylabel('Sales in $')
plt.show()
```

```python
## Q4 What is the most sold Product 

df = all_months_data[all_months_data['Order ID'].duplicated(keep = False)]

df['Grouped'] = df.groupby('Order ID') ['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)

```

```python

product_grp = all_months_data.groupby('Product')     
Qty_ordered = product_grp['Quantity Ordered'].sum()


Products = [Products for Products, df in product_grp]

plt.bar(Products, Qty_ordered)
plt.xticks(Products, rotation='vertical', size=9 )
plt.xlabel('Products')
plt.ylabel('No. of Sales')
plt.show()
```

```python
#answering the why

price = all_months_data.groupby('Product')['Price Each'].mean()

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(Products, Qty_ordered, color = 'g')
ax2.plot(Products, price, color = 'b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Qty_ordered', color = 'g')
ax2.set_ylabel('Price ($)', color = 'b')
ax1.set_xticklabels(Products, rotation = 'vertical', size = 7)

plt.show()
```
